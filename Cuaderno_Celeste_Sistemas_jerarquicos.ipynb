Indicamos los vectores de los CM pero solo los relativos

```python
m_Zul_Aa = 1
m_Zul_Ab = 0.01

r_Zul_A = np.array([0.1,0,0])
v_Zul_A = np.array([0,1.0,0])
```

```python
m_Zul_Aab = m_Zul_Aa + m_Zul_Ab
m_Zul_B = 1.0

r_Zul_AB = np.array([1.0,0.0,0.0])
v_Zul_AB = np.array([0.0,1.0,0.0])
```

Armar el sistema

Comenzamos definiendo el centro de masa de todo el sistema

```python
RCM_Zul_AB = np.array([0,0,0])
VCM_Zul_AB = np.array([0,0,0])
```

```python
#Este es el equivalente a la particula 2
r_Zul_B = RCM_Zul_AB - (m_Zul_Aab/(m_Zul_Aab + m_Zul_B)) * r_Zul_AB
v_Zul_B = VCM_Zul_AB - (m_Zul_Aab/(m_Zul_Aab + m_Zul_B)) * v_Zul_AB
    
#Este es el equivalente a la particula 1
r_Zul_A = RCM_Zul_AB + (m_Zul_B/(m_Zul_Aab + m_Zul_B)) * r_Zul_AB
v_Zul_A = VCM_Zul_AB + (m_Zul_B/(m_Zul_Aab + m_Zul_B)) * v_Zul_AB
```

En  el siguiente nivel del árbol:

```python
#Particula 1
r_Zul_Aa = RCM_Zul_A + (m_Zul_Ab/(m_Zul_Aab)) * r_Zul_A
v_Zul_Aa = VCM_Zul_A + (m_Zul_Ab/(m_Zul_Aab)) * v_Zul_A


#Particula 2
r_Zul_Ab = RCM_Zul_A + (m_Zul_Aa/(m_Zul_Aab)) * r_Zul_A
v_Zul_Ab = VCM_Zul_A + (m_Zul_Aa/(m_Zul_Aab)) * v_Zul_A
```
