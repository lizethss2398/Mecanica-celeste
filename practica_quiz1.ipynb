```python
import numpy as np

import matplotlib.pyplot as plt

# SOLUCIÓN DE ECUACIÓN DIFERENCIAL DE SEGUNDO ORDEN POR MÉTODO DE EULER
# Ecuación: d²y/dt² = -y/t²
# Condiciones iniciales: y(1) = 1, dy/dt(1) = 0

print("="*70)
print("TEORÍA DEL MÉTODO DE EULER PARA EDO DE SEGUNDO ORDEN")
print("="*70)
print("""
1. TRANSFORMACIÓN A SISTEMA DE PRIMER ORDEN:
    Una EDO de segundo orden se convierte en un sistema de 2 EDOs de primer orden
    
    Sea: y' = dy/dt = v
    Entonces: d²y/dt² = dv/dt = -y/t²
    
    Sistema resultante:
    ├─ dy/dt = v
    └─ dv/dt = -y/t²

2. MÉTODO DE EULER:
    Para una EDO: dy/dt = f(t, y)
    La aproximación es: y(t + h) ≈ y(t) + h · f(t, y)
    
    En nuestro caso:
    ├─ y_{n+1} = y_n + h · v_n
    └─ v_{n+1} = v_n + h · (-y_n / t_n²)
    
    donde:
    - h = paso de integración (tamaño del intervalo)
    - n = índice actual
    - t_n, y_n, v_n = valores en el paso actual
""")

# PARÁMETROS DE LA SOLUCIÓN
print("\n" + "="*70)
print("CONFIGURACIÓN")
print("="*70)

t_inicial = 1.0      # Tiempo inicial
t_final = 10.0       # Tiempo final
h = 0.01            # Tamaño del paso (h pequeño → mayor precisión)
y_inicial = 1.0     # Condición inicial: y(1) = 1
v_inicial = 0.0     # Condición inicial: dy/dt(1) = 0

print(f"Intervalo de integración: t ∈ [{t_inicial}, {t_final}]")
print(f"Tamaño del paso: h = {h}")
print(f"Condiciones iniciales: y(1) = {y_inicial}, y'(1) = {v_inicial}")

# INICIALIZACIÓN DE ARRAYS
num_pasos = int((t_final - t_inicial) / h) + 1
t = np.zeros(num_pasos)
y = np.zeros(num_pasos)
v = np.zeros(num_pasos)  # v = dy/dt

# VALORES INICIALES
t[0] = t_inicial
y[0] = y_inicial
v[0] = v_inicial

print("\n" + "="*70)
print("PROCESO ITERATIVO (PRIMEROS 5 PASOS)")
print("="*70)

# MÉTODO DE EULER APLICADO PASO A PASO
for n in range(num_pasos - 1):
     # Evaluación de la derivada en el punto actual
     dvdt = -y[n] / (t[n]**2)  # Segundo miembro de dv/dt = -y/t²
     
     # Actualización de variables (método de Euler)
     y[n+1] = y[n] + h * v[n]
     v[n+1] = v[n] + h * dvdt
     t[n+1] = t[n] + h
     
     # Mostrar primeros 5 pasos para verificación
     if n < 5:
          print(f"\nPaso {n}:")
          print(f"  t_{n} = {t[n]:.4f}")
          print(f"  y_{n} = {y[n]:.6f}, v_{n} = {v[n]:.6f}")
          print(f"  dv/dt|_n = -y/t² = -{y[n]:.6f}/{t[n]:.4f}² = {dvdt:.6f}")
          print(f"  y_{n+1} = {y[n]:.6f} + {h:.4f} × {v[n]:.6f} = {y[n+1]:.6f}")
          print(f"  v_{n+1} = {v[n]:.6f} + {h:.4f} × {dvdt:.6f} = {v[n+1]:.6f}")

print("\n" + "="*70)
print("GRÁFICAS DE RESULTADOS")
print("="*70)

# VISUALIZACIÓN
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Gráfica 1: Solución y(t)
axes[0].plot(t, y, 'b-', linewidth=2, label='y(t)')
axes[0].scatter([t_inicial], [y_inicial], color='red', s=100, zorder=5, label='Condición inicial')
axes[0].set_xlabel('t', fontsize=12)
axes[0].set_ylabel('y', fontsize=12)
axes[0].set_title('Solución: y(t)', fontsize=14, fontweight='bold')
axes[0].grid(True, alpha=0.3)
axes[0].legend(fontsize=11)

# Gráfica 2: Derivada v(t) = dy/dt
axes[1].plot(t, v, 'g-', linewidth=2, label="v(t) = dy/dt")
axes[1].scatter([t_inicial], [v_inicial], color='red', s=100, zorder=5, label='Condición inicial')
axes[1].set_xlabel('t', fontsize=12)
axes[1].set_ylabel("dy/dt", fontsize=12)
axes[1].set_title("Derivada: dy/dt", fontsize=14, fontweight='bold')
axes[1].grid(True, alpha=0.3)
axes[1].legend(fontsize=11)

plt.tight_layout()
plt.show()

# TABLA DE RESULTADOS (ÚLTIMOS 5 PUNTOS)
print("\nÚLTIMOS 5 PUNTOS DE LA SOLUCIÓN:")
print("-"*50)
print(f"{'t':>10} | {'y(t)':>15} | {'dy/dt':>15}")
print("-"*50)
for i in range(-5, 0):
     print(f"{t[i]:>10.4f} | {y[i]:>15.8f} | {v[i]:>15.8f}")
```
